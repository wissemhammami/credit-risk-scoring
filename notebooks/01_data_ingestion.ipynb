```python
# 01_data_ingestion.ipynb
# Purpose:
# - Load raw Home Credit datasets
# - Perform basic existence checks
# - Quick sanity check for missing values and target distribution

import pandas as pd
from pathlib import Path

# --------------------------------------------------
# Project paths (anchor to project root)
# --------------------------------------------------
BASE_DIR = Path.cwd().parent
RAW_DATA_DIR = BASE_DIR / "data" / "raw"

# --------------------------------------------------
# Raw files expected from Kaggle
# --------------------------------------------------
RAW_FILES = [
    "application_train.csv",
    "application_test.csv",
    "bureau.csv",
    "bureau_balance.csv",
    "credit_card_balance.csv",
    "installments_payments.csv",
    "POS_CASH_balance.csv",
    "previous_application.csv",
]

# --------------------------------------------------
# Load datasets
# --------------------------------------------------
datasets = {}
print("Loading datasets...\n")

for filename in RAW_FILES:
    path = RAW_DATA_DIR / filename

    if not path.exists():
        print(f"[MISSING] {filename}")
        continue

    key = filename.replace(".csv", "")
    datasets[key] = pd.read_csv(path)
    print(f"[LOADED] {filename} | shape={datasets[key].shape}")

# --------------------------------------------------
# Summary
# --------------------------------------------------
print("\nDatasets loaded:")
for name in datasets:
    print(f" - {name}: {datasets[name].shape}")

# --------------------------------------------------
# Quick sanity check for application_train
# --------------------------------------------------
if "application_train" in datasets:
    train_df = datasets["application_train"]

    # First 5 rows
    print("\nFirst 5 rows of application_train:\n")
    print(train_df.head(), "\n")

    # Missing values (only columns with missing values)
    missing_summary = train_df.isna().sum()
    print("Missing values in application_train:\n")
    print(missing_summary[missing_summary > 0], "\n")

    # Target distribution
    target_dist = train_df["TARGET"].value_counts(normalize=True).rename("proportion")
    print("Target distribution in application_train:\n")
    print(target_dist)
```
