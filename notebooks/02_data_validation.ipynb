```python
# 02_data_validation.ipynb
# Purpose:
# - Reload raw datasets
# - Perform basic sanity checks
# - Identify missing values, duplicates, and target distribution
# - No cleaning, no saving

import pandas as pd
from pathlib import Path

# --------------------------------------------------
# Project paths
# --------------------------------------------------
BASE_DIR = Path.cwd().parent
RAW_DATA_DIR = BASE_DIR / "data" / "raw"

# --------------------------------------------------
# Raw files expected
# --------------------------------------------------
RAW_FILES = [
    "application_train.csv",
    "application_test.csv",
    "bureau.csv",
    "bureau_balance.csv",
    "credit_card_balance.csv",
    "installments_payments.csv",
    "POS_CASH_balance.csv",
    "previous_application.csv",
]

# --------------------------------------------------
# Load datasets
# --------------------------------------------------
datasets = {}
print("Loading datasets...\n")

for filename in RAW_FILES:
    path = RAW_DATA_DIR / filename
    if path.exists():
        key = filename.replace(".csv", "")
        datasets[key] = pd.read_csv(path)
        print(f"[LOADED] {filename} | shape={datasets[key].shape}")
    else:
        print(f"[MISSING] {filename}")

# --------------------------------------------------
# Basic validation checks
# --------------------------------------------------
print("\nBasic validation summary:\n")
for name, df in datasets.items():
    missing_count = df.isna().sum()
    print(f"{name}:")
    print(f"  rows: {df.shape[0]}")
    print(f"  cols: {df.shape[1]}")
    # Only show columns with missing values
    if missing_count.sum() > 0:
        print(f"  missing values:\n{missing_count[missing_count > 0]}")
    else:
        print("  missing values: None")
    print(f"  duplicate rows: {df.duplicated().sum()}")
    print("-" * 40)

# --------------------------------------------------
# Target sanity check (train only)
# --------------------------------------------------
if "application_train" in datasets:
    train_df = datasets["application_train"]
    target_dist = train_df["TARGET"].value_counts(normalize=True).rename("proportion")
    print("\nTarget distribution in application_train:\n")
    print(target_dist)
```
