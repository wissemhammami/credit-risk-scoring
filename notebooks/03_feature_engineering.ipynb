```python
# 03_feature_engineering.ipynb:

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ------------------------------
# Ensure datasets are loaded
# ------------------------------
try:
    datasets
except NameError:
    raise ValueError("Run 01_data_ingestion.ipynb first to load datasets into 'datasets'.")

# ------------------------------
# Split train/test
# ------------------------------
train_df = datasets["application_train"].copy()
test_df = datasets["application_test"].copy()

y = train_df["TARGET"].copy()
X = train_df.drop(columns=["TARGET"])
X_test = test_df.copy()

# ------------------------------
# Keep only numeric columns
# ------------------------------
numeric_cols = X.select_dtypes(include=np.number).columns
X = X[numeric_cols]
X_test = X_test[numeric_cols]

# ------------------------------
# Fill missing values
# ------------------------------
X = X.fillna(X.mean())
X_test = X_test.fillna(X.mean())

# ------------------------------
# Standard scaling
# ------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)

# ------------------------------
# Train/Validation split
# ------------------------------
X_train_scaled, X_val_scaled, y_train, y_val = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# ------------------------------
# PCA for dimensionality reduction (95% variance)
# ------------------------------
pca = PCA(n_components=0.95, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_val_pca = pca.transform(X_val_scaled)
X_test_pca = pca.transform(X_test_scaled)

# ------------------------------
# Summary
# ------------------------------
print("Shapes after preprocessing:")
print(f"X_train_pca: {X_train_pca.shape}")
print(f"X_val_pca:   {X_val_pca.shape}")
print(f"X_test_pca:  {X_test_pca.shape}")
print(f"y_train:     {y_train.shape}")
print(f"y_val:       {y_val.shape}")
```
